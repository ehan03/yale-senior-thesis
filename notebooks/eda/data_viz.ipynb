```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid", context="paper")
```

```python
features_df = pd.read_pickle("../../data/features.pkl.xz")
features_df
```

```python
subset = features_df.loc[features_df["red_win"].notnull()].copy()
subset["red_win"] = subset["red_win"].astype(int)
subset
```

```python
subset["cumulative_significant_strikes_landed_per_second_diff"]
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
sns.histplot(data=subset, x="age_days_diff", hue="red_win", ax=ax)
ax.set_xlabel("Difference in age in days")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
sns.histplot(data=subset, x="avg_glicko_1_change_diff", hue="red_win", ax=ax)
ax.set_xlabel("Difference in average post-fight Glicko-1 rating change")
plt.show()
```

```python
cols = subset.columns
temp = [col for col in cols if "total_time_" in col]
temp
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
sns.histplot(data=subset, x="avg_total_time_fought_seconds_diff_diff", hue="red_win", ax=ax)
ax.set_xlabel("Difference in average difference in total time fought between fighters and their opponents")
plt.show()
```
